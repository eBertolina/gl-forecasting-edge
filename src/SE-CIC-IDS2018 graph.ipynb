```python
from pathlib import Path
import json
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import ast
import os

sim_numbers = [0, 1, 2]
full_dataset_v = []
sampled_dataset_v = []

for n in sim_numbers:
    path = Path(f"experiments/SE-CIC-IDS2018/{n} FullDataset unsampled/plots/testing_unified")
    with open(path / 'metrics_avg.json') as f:
        d = json.load(f)
    f.close()
    full_dataset_v.append(d[4])

    path_uns = Path(f"experiments/SE-CIC-IDS2018/{n} SubSampled/plots/testing_unified")
    with open(path_uns / 'metrics_avg.json') as f:
        d = json.load(f)
    f.close()
    sampled_dataset_v.append(d[4])

print(full_dataset_v)
print(sampled_dataset_v)
```

```python
import matplotlib.pyplot as plt

#xs = [x for x in range(len(sim_numbers))]

plt.figure(figsize=(10,8))
plt.xlabel("Simulazioni")
plt.ylabel("Value")
plt.xticks(sim_numbers)
#plt.plot(xs, accuracy_local_list)
plt.plot(
    full_dataset_v,
    label="Full Dataset F1 weighted",
    color="royalblue",
    linewidth=2,
)
plt.plot(
    sampled_dataset_v,
    label="Subsampled dataset F1 weighted",
    color="limegreen",
    linewidth=2,
)
plt.legend()
plt.savefig( Path(f"experiments/SE-CIC-IDS2018/") / "comparison_plot.jpg")
plt.show()
plt.close()
```
